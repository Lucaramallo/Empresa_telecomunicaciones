```python
import pandas as pd
```

```python
url = 'https://datosabiertos.enacom.gob.ar/rest/datastreams/255796/data.csv'
df_tipo_tecn_x_prov = pd.read_csv(url,encoding='UTF-8',sep=',',thousands='.')
```

```python
df_tipo_tecn_x_prov.isnull().sum()
```

```python
df_tipo_tecn_x_prov = df_tipo_tecn_x_prov.fillna(0)
```

```python
# Eliminar la última fila
df_tipo_tecn_x_prov = df_tipo_tecn_x_prov.drop(df_tipo_tecn_x_prov.index[-1])
```

```python
df_tipo_tecn_x_prov.describe()
```

```python
df_tipo_tecn_x_prov.describe(include='all')
```

```python
# Renombrar la columna 'Año' a 'anio'
df_tipo_tecn_x_prov.rename(columns={'Año': 'anio'}, inplace=True)
# Convertir todas las columnas a minúsculas utilizando rename() y lower()
df_tipo_tecn_x_prov = df_tipo_tecn_x_prov.rename(columns=str.lower)
```

```python
df_tipo_tecn_x_prov['anio'] = df_tipo_tecn_x_prov['anio'].str.replace('2019 \*', '2019')
df_tipo_tecn_x_prov['trimestre'] = df_tipo_tecn_x_prov['trimestre'].str.replace('3 \*', '3')
df_tipo_tecn_x_prov['trimestre'] = df_tipo_tecn_x_prov['trimestre'].str.replace('2 \*', '2')
df_tipo_tecn_x_prov['trimestre'] = df_tipo_tecn_x_prov['trimestre'].str.replace('1 \*', '1')
# Filtrar y mostrar las filas con "2019" en la columna "anio"
df_filtered = df_tipo_tecn_x_prov[df_tipo_tecn_x_prov['anio'] == '2019']
```

```python
df_filtered.head(5)
```

```python
df_filtered.tail(5)
```

```python
df_tipo_tecn_x_prov['anio_tri'] = df_tipo_tecn_x_prov['anio'].astype(str) + df_tipo_tecn_x_prov['trimestre'].astype(str)
```

```python
# Situar la columna 'anio_tri' como la primera columna
df_tipo_tecn_x_prov = df_tipo_tecn_x_prov[['anio_tri'] + df_tipo_tecn_x_prov.columns[:-1].tolist()]
```

```python
# ordenar el DataFrame por año y trimestre
df_tipo_tecn_x_prov = df_tipo_tecn_x_prov.sort_values(by=["anio_tri"])
```

```python
# Eliminar la columna "total"
df_tipo_tecn_x_prov = df_tipo_tecn_x_prov.drop(columns=['total'])
```

```python
import numpy as np

# Calcular la suma de las columnas restantes y redondear
df_tipo_tecn_x_prov['totales'] = df_tipo_tecn_x_prov.iloc[:, 4:].sum(axis=1).round().astype(np.int64)
```

```python
# Convertir los valores de tipo float a int64
column_start_index = df_tipo_tecn_x_prov.columns.get_loc('adsl')
df_tipo_tecn_x_prov.iloc[:, column_start_index:] = df_tipo_tecn_x_prov.iloc[:, column_start_index:].astype('int64')
```
