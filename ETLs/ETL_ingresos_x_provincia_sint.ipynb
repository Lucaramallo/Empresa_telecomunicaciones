# Pivot:

##### El pivot table de los accesos por cada100 hogares , denota la participacion de cada provincia en el total del pais, por otra parte teniendo los ingresos acomodados en la misma estructura de tiempos, pero estos ultimos solo a nivel anual... quiere decir que normalizando la tabla 1 y multiplicando por la 2 obtenemos el ingreso monetario que aporto cada provincia por cada acceso cada 100 hogares.


```python
# Hacer la transpuesta o pivot table
df_pivot_df_pen_int_fij_x100h = df_pen_int_fij_x100h.pivot(index='provincia', columns=['anio_tri'], values=['accesos por cada 100 hogares'])
```

```python
df_pivot_df_pen_int_fij_x100h # agrupado por provincia, accesos x cada 100 h. 24 rows
```

```python
from sklearn.preprocessing import normalize

# Normalizar el DataFrame
df_normalized_df_pivot_df_pen_int_fij_x100h = normalize(df_pivot_df_pen_int_fij_x100h, axis=1)

# Crear un nuevo DataFrame con los datos normalizados y las mismas columnas e índices
df_normalized_df_pivot_df_pen_int_fij_x100h = pd.DataFrame(df_normalized_df_pivot_df_pen_int_fij_x100h, columns=df_pivot_df_pen_int_fij_x100h.columns, index=df_pivot_df_pen_int_fij_x100h.index)
```

```python
df_normalized_df_pivot_df_pen_int_fij_x100h.head(5)
```

```python
# Revertir la tabla pivotada
df_pivot_df_pen_int_fij_x100h_normalized_revertido = df_normalized_df_pivot_df_pen_int_fij_x100h.stack().reset_index()
df_pivot_df_pen_int_fij_x100h_normalized_revertido.columns = ['provincia', 'anio_tri', 'accesos por cada 100 hogares']

df_pivot_df_pen_int_fij_x100h_normalized_revertido
```

# Genero matriz plana de ingresos x periodo

```python
# Suponiendo que ya tienes el dataframe "df_pivot_df_ingresos_inter_fijo"

# Obtén el número de filas y columnas del dataframe
num_filas, num_columnas = df_pivot_df_ingresos_inter_fijo.shape

# Itera sobre cada fila del dataframe
for i in range(num_filas):
    fila = df_pivot_df_ingresos_inter_fijo.iloc[i, :]  # Obtiene la fila actual
    
    # Encuentra el único valor mayor a cero en la fila
    valores_fila_positivos = fila[fila > 0]
    if len(valores_fila_positivos) == 1:
        valor_mayor = valores_fila_positivos.iloc[0]
    else:
        valor_mayor = 0.0
    
    # Actualiza todos los valores de la fila con el valor mayor
    df_pivot_df_ingresos_inter_fijo.iloc[i, :] = valor_mayor

# Muestra el dataframe actualizado
df_pivot_df_ingresos_inter_fijo.head(5)
```

#### Extraigo la mascara de valores unicamente, ya acomodados (operacion sin index):


```python
columna_deseada = df_pivot_df_ingresos_inter_fijo.columns[0]  # Obtener el nombre de la primera columna
df_valores = pd.DataFrame({
    'anio_tri': df_pivot_df_ingresos_inter_fijo.index,
    columna_deseada: df_pivot_df_ingresos_inter_fijo[columna_deseada].values
})

# Establecer "anio_tri" como el nuevo índice
df_valores.set_index('anio_tri', inplace=True)

# Mostrar el nuevo DataFrame con la columna de valores y el índice mantenido
df_valores.head(5)

```

```python
# Transponer el DataFrame
df_pivot = df_valores.transpose()

# Resetear el índice
df_pivot.reset_index(drop=True, inplace=True)

# Mostrar el DataFrame pivoteado
df_pivot
```

```python
df_resultado = pd.DataFrame()

for index, row in df_normalized_df_pivot_df_pen_int_fij_x100h.iterrows():
    resultado_fila = df_pivot.mul(row)
    df_resultado = df_resultado.append(resultado_fila)

df_resultado.reset_index(drop=True, inplace=True)
```
