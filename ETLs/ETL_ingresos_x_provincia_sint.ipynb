```python
# Codigo reciclado. intento 2
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Create a new dataframe with the data from the last three months and the next three months for each province
df_regresion = df_resultado.iloc[:, -6:]

# Define X and y
X = np.array(df_regresion.iloc[:, :-3])
y = np.array(df_regresion.iloc[:, -3:])

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=7)

# Create an instance of RandomForestRegressor and fit the model to the training data
reg = RandomForestRegressor()
reg.fit(X_train, y_train)

# Make predictions for the next three months using the test data
predictions = reg.predict(X_test)

# Print the predictions
print(predictions)

# Calculate evaluation metrics on the test data
mae = mean_absolute_error(y_test, predictions)
mse = mean_squared_error(y_test, predictions)

# Print the evaluation metrics
print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)

# Perform cross-validation to get a more robust evaluation of the model's performance
cv_scores = cross_val_score(reg, X, y, scoring='neg_mean_squared_error', cv=5)
rmse_cv = np.sqrt(-cv_scores)

# Print the cross-validated RMSE scores
print("Cross-Validated RMSE Scores:", rmse_cv)

# Fit the model to the entire data and make predictions for the next three months
reg.fit(X, y)
predictions = reg.predict(X)

# Add the three forecast columns to the df_resultado dataframe
columnas_pronostico = [20224, 20231, 20232] # deberia de modificarlo para que obtenga el ultimo periodo y calcule el que corresponda... agregar despues!
for i, col in enumerate(columnas_pronostico):
    df_resultado[col] = predictions[:, i]

# Print the updated dataframe
print(df_resultado)
```
