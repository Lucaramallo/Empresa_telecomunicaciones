```python
import pandas as pd
```

```python
df = pd.read_csv('../CSV_Cleaned/ingresos_x_provincia_x_periodo.csv')
```

```python

# Obtener valores únicos de la columna "provincia"
provincias_unicas = df['provincia'].unique()
```
