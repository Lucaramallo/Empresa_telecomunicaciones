```python
import pandas as pd
```

```python
df_rangos_de_vel = pd.read_csv('../CSV data importada/CSV Accesos a internet/Acceso a Internet Fijo por rangos de velocidad de bajada y provincia.csv',encoding='UTF-8',sep=',',thousands='.')
```

```python
df_rangos_de_vel = df_rangos_de_vel.fillna(0)
df_rangos_de_vel['OTROS'] = df_rangos_de_vel['OTROS'].astype('Int64')
```

```python
# Renombrar la columna 'Año' a 'anio'
df_rangos_de_vel.rename(columns={'Año': 'anio'}, inplace=True)
# Convertir todas las columnas a minúsculas utilizando rename() y lower()
df_rangos_de_vel = df_rangos_de_vel.rename(columns=str.lower)
```

```python
df_rangos_de_vel['anio_tri'] = df_rangos_de_vel['anio'].astype(str) + df_rangos_de_vel['trimestre'].astype(str)
```

```python
# Situar la columna 'anio_tri' como la primera columna
df_rangos_de_vel = df_rangos_de_vel[['anio_tri'] + df_rangos_de_vel.columns[:-1].tolist()]
```

```python
sns.pairplot(df_rangos_de_vel)
```

```python
# ordenar el DataFrame por año y trimestre
df_rangos_de_vel = df_rangos_de_vel.sort_values(by=["anio_tri"])
```

```python
# Eliminar la columna "total"
df_rangos_de_vel = df_rangos_de_vel.drop(columns=['total'])
```

```python
import numpy as np

# Calcular la suma de las columnas restantes y redondear
df_rangos_de_vel['totales'] = df_rangos_de_vel.iloc[:, 4:].sum(axis=1).round().astype(np.int64)
```
