# ETL Penetración de Telefonia movil (accesos por cada 100 hogares):

```python
import pandas as pd
#Penetración nacional de la telefonía móvil (accesos por cada 100 habitantes)
```

```python
url = 'https://datosabiertos.enacom.gob.ar/rest/datastreams/288818/data.csv'
df_pen_tel_movil_x100h = pd.read_csv(url, encoding='UTF-8', sep=',', decimal=',' )
```

```python
# Renombrar la columna 'Año' a 'anio'
df_pen_tel_movil_x100h.rename(columns={'Año': 'anio'}, inplace=True)
# Convertir todas las columnas a minúsculas utilizando rename() y lower()
df_pen_tel_movil_x100h = df_pen_tel_movil_x100h.rename(columns=str.lower)
```

```python
df_pen_tel_movil_x100h.head()
```

```python
df_pen_tel_movil_x100h['anio_tri'] = df_pen_tel_movil_x100h['anio'].astype(str) + df_pen_tel_movil_x100h['trimestre'].astype(str)
```

```python
# Situar la columna 'anio_tri' como la primera columna
df_pen_tel_movil_x100h = df_pen_tel_movil_x100h[['anio_tri'] + df_pen_tel_movil_x100h.columns[:-1].tolist()]
```

```python
df_pen_tel_movil_x100h.head()
```

```python
# ordenar el DataFrame por año y trimestre
df_pen_tel_movil_x100h = df_pen_tel_movil_x100h.sort_values(by=["anio_tri"])
```

```python

# calcular la variación porcentual entre los valores de accesos de un trimestre a otro
df_pen_tel_movil_x100h["variacion de los accesos por cada 100 habitantes"] = df_pen_tel_movil_x100h["accesos por cada 100 hab"].pct_change()

# rellenar los valores nulos con ceros en la columna "variacion"
df_pen_tel_movil_x100h["variacion de los accesos por cada 100 habitantes"] = df_pen_tel_movil_x100h["variacion de los accesos por cada 100 habitantes"].fillna(0)

# mostrar el resultado
df_pen_tel_movil_x100h.head(5)
```
