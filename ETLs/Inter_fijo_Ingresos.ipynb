```python
import pandas as pd
```

```python
df_ingresos_inter_fijo = pd.read_csv('../CSV data importada/CSV Accesos a internet/Internet Ingresos.csv',thousands='.', encoding='UTF-8', sep=',')
```

```python
df_ingresos_inter_fijo.describe(include='all')
```

```python
# Renombrar la columna 'Año' a 'anio'
df_ingresos_inter_fijo.rename(columns={'Año': 'anio'}, inplace=True)
# Convertir todas las columnas a minúsculas utilizando rename() y lower()
df_ingresos_inter_fijo = df_ingresos_inter_fijo.rename(columns=str.lower)
```

```python
df_ingresos_inter_fijo.head()
```

```python
df_ingresos_inter_fijo['anio_tri'] = df_ingresos_inter_fijo['anio'].astype(str) + df_ingresos_inter_fijo['trimestre'].astype(str)
```

```python
# Situar la columna 'anio_tri' como la primera columna
df_ingresos_inter_fijo = df_ingresos_inter_fijo[['anio_tri'] + df_ingresos_inter_fijo.columns[:-1].tolist()]
```

```python
df_ingresos_inter_fijo.head()
```

```python
import seaborn as sns
```

```python
sns.pairplot(df_ingresos_inter_fijo,hue='anio_tri')
```

```python
df_ingresos_inter_fijo.head(5)
```

```python
df_ingresos_inter_fijo = df_ingresos_inter_fijo.sort_values(by=['anio_tri'])
```

```python
df_ingresos_inter_fijo
```

```python
# Calcular la variación porcentual
df_ingresos_inter_fijo['variacion trimestral ingresos (miles de pesos)'] = df_ingresos_inter_fijo['ingresos (miles de pesos)'].pct_change()

# Reordenar las columnas para que la nueva columna quede junto a "trimestre" y "ingresos (miles de pesos)"
df_ingresos_inter_fijo = df_ingresos_inter_fijo[['anio_tri', 'anio', 'trimestre', 'variacion trimestral ingresos (miles de pesos)', 'ingresos (miles de pesos)', 'periodo']]
```
